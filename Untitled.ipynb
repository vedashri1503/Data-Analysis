```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("tesla-stock-price.csv")
print(df.head()) 
```

```python
plt.scatter(df.date,df.volume)
plt.show()
```

```python
plt.scatter(df.date,df.volume)
sns.set_style("dark")
plt.show()
```

```python
plt.scatter(df.date,df.volume)
sns.set_style("whitegrid")
sns.set_context("notebook")  
plt.show()
```

```python
sns.displot(df['high'],kde=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='high', y='volume', data=df, hue='high', palette='PuRd', errorbar=None)
plt.xticks(rotation=45) 
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='high', y='volume', data=df, palette='PuRd', errorbar=None)  # Swap x and y for horizontal bars
plt.show()  # No need for plt.legend()
```

```python
sns.set_context('paper')

sns.countplot(x='high', hue='volume',data=df,palette='magma')
plt.title('Survivors')
plt.show()
```

```python
import seaborn as sns

sns.jointplot(x='high', y='volume', data=df, kind='scatter')  # or kind='hist'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="high", y="weather_main", data=df)
plt.show()
```
